# Initial imports for creating a neural network

```python
import keras as kr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make the plots bigger 
plt.rcParams['figure.figsize']=(10,10)
```

# Single Neuron Network

```python
# These three commands are always uses to create neural networks for now
# Creates a new Neural Network
m = kr.models.Sequential()

# add a single neuron layer with a weight of 1 and a bias of 0
# First parameter: The number of layers in the neural network
# Second parameter: The number of inputs taken by the neural network, in this case it's 1
# Third parameter: Do not tamper with the outputs of this layer
# Fourth parameter: sets all of the weights on the first layer of the network to 1
# Fifth parameter: sets all of the bias on the first layer of the network to 0
m.add(kr.layers.Dense(1,input_dim=1, activation='linear', kernel_initializer='ones', bias_initializer='zeros'))

# Compiles the model
m.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
# Create input values
x = np.arange(0,10,1)

# Run each x(value) through the neural network
y = m.predict(x)

# Plot the values on a graph
# Linear Equation
plt.plot(x,y, 'rx')
```

```python

# Creates a new Neural Network
m = kr.models.Sequential()

# add a single neuron layer with a weight of 1 and a bias of 0

m.add(kr.layers.Dense(1,input_dim=1, activation='linear', kernel_initializer=kr.initializers.Constant(value=2), bias_initializer=kr.initializers.Constant(value=1)))

# Compiles the model
m.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
# Create input values
x = np.arange(0,10,1)

# Run each x(value) through the neural network
y = m.predict(x)

# Plot the values on a graph
# Linear Equation
plt.plot(x,y, 'go')
```

# Two Neuron Network

```python
# Need at least three neurons but one can be a dummy neuron
m = kr.models.Sequential()

# Add two neurons in a single layer
m.add(kr.layers.Dense(2, input_dim=1, activation='linear'))

# Add a single neuron in a single layer with weight 1 and bias 0
# Dont need to put anything in as input_dim because it knows its inputs 
# are from the two neurons on the previous layer 
m.add(kr.layers.Dense(1,input_dim=1, activation='linear', kernel_initializer=kr.initializers.Constant(value=1), bias_initializer=kr.initializers.Constant(value=0)))

# Set the weight and bias of the two neurons 
m.layers[0].set_weights([np.matrix([2,3]), np.array([5,-3])])

# Compiles the model
m.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
# Create input values
x = np.arange(0,10,1)

# Run each x(value) through the neural network
y = m.predict(x)

# Plot the values on a graph
# Linear Equation
plt.plot(x,y, 'b.')
```
